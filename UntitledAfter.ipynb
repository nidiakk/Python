#Chap1 變數的應用
Q:邊長為3,4,5的三角形，其面積大小為何?
<br>海龍公式:
<br>s=(a+b+c)/2
<br>sqrt(s(s-a)(s-b)(s-c))

```
import math
a,b,c=3,4,5
s=(a+b+c)/2
d=math.sqrt(s*(s-a)*(s-b)*(s-c))
print(d)
```

##for 迴圈

```
#1+2+3....+10總和
sum=0
for count in range(1,11,1):#從1開始，間隔1，不包含11
  sum+=count ##空四個空格/一個tab鍵就像是別的語法的{}括弧，用以區隔。
print("1+2+3...+10=",sum)
```

```
#range另一種寫法，較常這樣用，直接就打範圍是11
#1+2+3....+10總和
sum=0
for count in range(11):#從1開始，間隔1，不包含11
  sum+=count ##空四個空格/一個tab鍵就像是別的語法的{}括弧，用以區隔。
print("1+2+3...+10=",sum)
```
